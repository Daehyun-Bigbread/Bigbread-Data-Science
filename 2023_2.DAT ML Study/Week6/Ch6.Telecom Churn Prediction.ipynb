### EDA & Prediction

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # For creating plots
import matplotlib.ticker as mtick # For specifying the axes tick format 
import matplotlib.pyplot as plt

sns.set(style = 'white')

# Input data files are available in the "../input/" directory.
```

```python
telecom_cust = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

### 결측치 확인

```python
# Checking the data types of all the columns
telecom_cust.dtypes
```

```python
# Converting Total Charges to a numerical data type.
telecom_cust.TotalCharges = pd.to_numeric(telecom_cust.TotalCharges, errors='coerce')
telecom_cust.isnull().sum()
```

```python
#Removing missing values 
telecom_cust.dropna(inplace = True)
#Remove customer IDs from the data set
df2 = telecom_cust.iloc[:,1:]
#Convertin the predictor variable in a binary numeric variable
df2['Churn'].replace(to_replace='Yes', value=1, inplace=True)
df2['Churn'].replace(to_replace='No',  value=0, inplace=True)

#Let's convert all the categorical variables into dummy variables
df_dummies = pd.get_dummies(df2)
df_dummies.head()
```

```python
#Get Correlation of "Churn" with other variables:
plt.figure(figsize=(15,8))
df_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```

### Logistic Regression

```python
# 로지스틱 회귀
# 더미 변수를 생성한 데이터 프레임을 사용할 것입니다.
y = df_dummies['Churn'].values  # 종속 변수로 'Churn'을 사용합니다.
X = df_dummies.drop(columns=['Churn'])  # 'Churn' 열을 제외한 모든 열을 독립 변수로 사용합니다.

# 모든 변수를 0에서 1 사이의 범위로 스케일링합니다.
from sklearn.preprocessing import MinMaxScaler
features = X.columns.values
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X))
X.columns = features  # 스케일링 후, 열 이름을 원래 변수 이름으로 다시 설정합니다.
```

로지스틱 회귀 분석에서 변수의 크기를 0에서 1의 범위 내에 있도록 조정하는 것이 중요합니다. 이를 통해 정확도가 79.7%에서 80.7%로 향상되었습니다. 또한 아래에서 확인할 수 있듯이 변수의 중요도는 Random Forest 알고리즘과 위에서 수행한 EDA에서도 일치합니다.

```python
# Create Train & Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Running logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics
prediction_test = model.predict(X_test)
# Print the prediction accuracy
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
# To get the weights of all the variables
weights = pd.Series(model.coef_[0],
                 index=X.columns.values)
print (weights.sort_values(ascending = False)[:10].plot(kind='bar'))
```

```python
print(weights.sort_values(ascending = False)[-10:].plot(kind='bar'))
```

우리는 우리가 예측한 변수(Churn)와 음의 관계를 가지는 반면에 양의 관계를 가지는 변수를 볼 수 있습니다. 음의 관계는 그 변수와 함께 Churn의 가능성이 감소한다는 것을 의미합니다. 아래 몇 가지 흥미로운 특징을 요약해 보겠습니다:

우리가 EDA에서 보았듯이, 2개월 계약을 하는 것은 이직의 기회를 줄여줍니다. 2개월 계약과 재직기간은 로지스틱 회귀분석에서 예측한 것처럼 가장 부정적인 관계를 가지고 있습니다
DSL 인터넷 서비스를 이용하면 춘절 확률도 줄어듭니다
마지막으로, 총 요금, 월 계약, 광섬유 인터넷 서비스 및 연공서열은 더 높은 이탈률로 이어질 수 있습니다. 이것은 흥미로운 점인데, 광섬유 서비스가 더 빠르긴 하지만 고객들은 이 때문에 이탈할 가능성이 있기 때문입니다. 왜 이런 일이 일어나는지 더 잘 알아보기 위해 더 많은 탐색이 필요하다고 생각합니다.

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 데이터를 학습용과 테스트용으로 분리합니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

# RandomForestClassifier 모델을 생성하고 학습합니다.
model_rf = RandomForestClassifier(n_estimators=1000, oob_score=True, n_jobs=-1,
                                  random_state=50, max_features="log2",
                                  max_leaf_nodes=30)
model_rf.fit(X_train, y_train)

# 모델을 사용하여 테스트 데이터에 대한 예측을 수행합니다.
prediction_test = model_rf.predict(X_test)

# 정확도를 계산하고 출력합니다.
accuracy = metrics.accuracy_score(y_test, prediction_test)
print(accuracy)
```

```python
importances = model_rf.feature_importances_
weights = pd.Series(importances,
                 index=X.columns.values)
weights.sort_values()[-10:].plot(kind = 'barh')
```

랜덤 포레스트 알고리즘부터 월별 계약, 보유 기간 및 총부담금은 이탈을 예측하는 가장 중요한 예측 변수입니다.
랜덤 포레스트의 결과는 로지스틱 회귀 분석의 결과와 매우 유사하며 EDA에서 기대했던 것과 일치합니다

여기서 발생하는 confusion_matrix 함수는 모델의 예측과 실제 레이블을 비교하여 다음 네 가지 결과를 계산합니다:

True Positive (TP): 모델이 정확하게 양성 클래스를 예측한 경우.
True Negative (TN): 모델이 정확하게 음성 클래스를 예측한 경우.
False Positive (FP): 모델이 실제로는 음성 클래스인데 양성 클래스로 잘못 예측한 경우 (오검출).
False Negative (FN): 모델이 실제로는 양성 클래스인데 음성 클래스로 잘못 예측한 경우 (누락).
